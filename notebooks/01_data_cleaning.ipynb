```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/raw/my_activites.csv")
```

```python
df.head()
df.info()
df.describe()
```

```python
df.rename(columns={
    "Data da atividade": "activity_date",
    "Tipo de atividade": "activity_type",
    "Distância": "distance_km",
    "Tempo de movimentação": "moving_time",
    "Velocidade média": "avg_speed",
    "Frequência cardíaca máxima": "max_hr",
    "Ganho de elevação": "elevation_gain",
    "Esforço relativo": "relative_effort",
    "Nome da atividade" : "activity_name",
    "Tempo decorrido" : "elapsed_time",
    "Velocidade máx." : "max_speed",
    "Ganho de elevação" : "elevation_gain",
    "Perda de elevação" : "elevation_loss",
    


}, inplace=True)
```

```python
months = {
    "jan": "Jan",
    "fev": "Feb",
    "mar": "Mar",
    "abr": "Apr",
    "mai": "May",
    "jun": "Jun",
    "jul": "Jul",
    "ago": "Aug",
    "set": "Sep",
    "out": "Oct",
    "nov": "Nov",
    "dez": "Dec"
}

# convertir en string
df["activity_date"] = df["activity_date"].astype(str)

# remplacer les mois
for pt, en in months.items():
    df["activity_date"] = df["activity_date"].str.replace(pt, en)

# nettoyer
df["activity_date"] = df["activity_date"].str.replace(" de ", " ")

# convertir en datetime
df["activity_date"] = pd.to_datetime(df["activity_date"], errors="coerce")

df["activity_date"].head()
```

```python
df["distance_km"] = pd.to_numeric(df["distance_km"], errors="coerce")
df["distance_km"].head()
df["distance_km"].dtype

df = df[df["distance_km"] > 0.5]
df = df.dropna(subset=["distance_km"])
```

```python
df["pace_min_km"] = (df["moving_time"] / 60) / df["distance_km"]
df["pace_min_km"].head
```

```python
df["week"] = df["activity_date"].dt.isocalendar().week
df["month"] = df["activity_date"].dt.month
df["year"] = df["activity_date"].dt.year
df["day_of_week"] = df["activity_date"].dt.day_name()
```

```python
df["training_load"] = df["distance_km"] * df["relative_effort"]
```

```python
columns_to_keep = [
    "activity_date",
    "activity_name",
    "activity_type",
    "distance_km",
    "moving_time",
    "elapsed_time",
    "avg_speed",
    "max_speed",
    "pace_min_km",
    "max_hr",
    "relative_effort",
    "elevation_gain",
    "elevation_loss",
    "year",
    "month",
    "week",
    "day_of_week",
    "training_load"
]

df = df[columns_to_keep]
```

```python
df.to_csv("../data/processed/clean_activities.csv", index=False)
```
